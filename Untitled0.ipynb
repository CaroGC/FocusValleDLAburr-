```python
import numpy as np
import time #para medir la eficiencia

#Creamos dos vectores aleatorios
tamano=10000
np.random.seed(0)
v1=np.random.random(tamano)
v2=np.random.random(tamano)

#Usaremos bucles anidados para implementar el producto vectorial en python purto
instinicial=time.process_time()
productopython=np.zeros((tamano,tamano))#producto vectorial
for i in range(tamano): #bucles aninados
    for j in range(tamano):
      productopython[i][j]=v1[i]*v2[j]
pythontime=time.process_time()-instinicial #medir el tiempo

#Producto vectorial con numpy
instinicial=time.process_time()
productoNumpy=np.outer(v1,v2)#outer de numpy es más eficiente
numpytime=time.process_time()-instinicial

#Evaluacion resultados
if (productoNumpy==productopython).all():
  print("Los resultados son iguales")
else:
  print("Los resultados son diferentes")

print("Tiempo de python",pythontime)
print("Tiempo de numpy",numpytime)
print("Mejora con Numpy",pythontime/numpytime)#Mejora relativa entre ejecución e implementación
```

```python
import matplotlib.pyplot as plt
import numpy as np

def identidad(z): #funcion identidad y=x
  return z
def d_identidad(z): #derivada de funcion identidad
  return 1

def visualizar(y1,y2,z):
  plt.plot(z,y1,label="$identidad$",linewidth=0.5)
  plt.plot(z,y2,label="$d_identidad$",linestyle="--",linewidth=0.8)
  plt.xlabel("z")
  plt.ylabel("y")
  plt.title("y vs z")
  plt.legend()
  plt.show()

z=np.linspace(-5,5,50)
#para convertir en arreglos
y1=np.zeros(len(z))
y2=np.zeros(len(z))
y1=identidad(z)
y2[:]=d_identidad(z)
visualizar(y1,y2,z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
  return 1/(1+np.exp(-z))
def d_sigmoid(z):
  return (1-sigmoid(z))*sigmoid(z)
#Visualizar
def visualizar(y1,y2,z):
  plt.plot(z,y1,label="$Sigmoid$",linewidth=0.5)
  plt.plot(z,y2,label="$Derivada$",linestyle="--",linewidth=0.8)
  plt.xlabel("z")
  plt.ylabel("y")
  plt.title("y vs z")
  plt.legend()
  plt.show()
z=np.linspace(-5,5,50)

y1=np.zeros(len(z))
y2=np.zeros(len(z))
y1=sigmoid(z)
y2[:]=d_sigmoid(z)
visualizar(y1,y2,z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(z):
  return np.maximum(0,z)
def d_relu(z):
  return z>0
#Visualizar
def visualizar(y1,y2,z):
  plt.plot(z,y1,label="$Relu$",linewidth=0.5)
  plt.plot(z,y2,label="$Derivada$",linestyle="--",linewidth=0.8)
  plt.xlabel("z")
  plt.ylabel("y")
  plt.title("y vs z")
  plt.legend()
  plt.show()
z=np.linspace(-5,5,50)

y1=np.zeros(len(z))
y2=np.zeros(len(z))
y1=relu(z)
y2[:]=d_relu(z)
visualizar(y1,y2,z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def lrelu(a,z):
  return np.maximum(a*z,z)
def d_lrelu(a,z):
  return np.where(z>0,1,a)
#Visualizar
def visualizar(y1,y2,z):
  plt.plot(z,y1,label="$LReLU$",linewidth=0.5)
  plt.plot(z,y2,label="$Derivada$",linestyle="--",linewidth=0.8)
  plt.xlabel("z")
  plt.ylabel("y")
  plt.title("y vs z")
  plt.legend()
  plt.show()
z=np.linspace(-5,5,1000)

y1=np.zeros(len(z))
y2=np.zeros(len(z))
a=0.1
y1=lrelu(a,z)
y2[:]=d_lrelu(a,z)
visualizar(y1,y2,z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def tanh(z):
  return (2/(1+np.exp(-2*z)))-1
def d_tanh(z):
  return (1-tanh(z)*tanh(z))
#Visualizar
def visualizar(y1,y2,z):
  plt.plot(z,y1,label="$Tanh$",linewidth=0.5)
  plt.plot(z,y2,label="$Derivada$",linestyle="--",linewidth=0.8)
  plt.xlabel("z")
  plt.ylabel("y")
  plt.title("y vs z")
  plt.legend()
  plt.show()
z=np.linspace(-5,5,50)

y1=np.zeros(len(z))
y2=np.zeros(len(z))
y1=tanh(z)
y2[:]=d_tanh(z)
visualizar(y1,y2,z)
```
